```python
# Import all SMOTE variants for comprehensive comparison
from imblearn.over_sampling import (
    SMOTE, 
    BorderlineSMOTE, 
    SVMSMOTE, 
    ADASYN
)
from sklearn.model_selection import train_test_split
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def preprocess_data(df, target_col):
    print("=" * 50)
    print("DATA PREPROCESSING")
    print("=" * 50)

    # Create a copy
    data = df.copy()

    # Separate features and target
    X = data.drop(target_col, axis=1)
    y = data[target_col]
    print("Original class distribution:", Counter(y))


    # Handle missing values
    print("Handling missing values...")
    for col in X.columns:
        if X[col].dtype == 'object':
            # For categorical: fill with mode
            X[col] = X[col].fillna(X[col].mode()[0])
        else:
            # For numerical: fill with median
            X[col] = X[col].fillna(X[col].median())

    # Encode categorical variables
    print("Encoding categorical variables...")
    label_encoders = {}
    for col in X.columns:
        if X[col].dtype == 'object':
            le = LabelEncoder()
            X[col] = le.fit_transform(X[col].astype(str))
            label_encoders[col] = le

    # Encode target variable if it's categorical
    target_encoder = None
    if y.dtype == 'object':
        target_encoder = LabelEncoder()
        y = target_encoder.fit_transform(y)

    print(f"Preprocessed data shape: {X.shape}")
    print(f"Target classes: {np.unique(y)}")

    return X, y, label_encoders, target_encoder

X, y, encoders, target_encoder = preprocess_data(df, target_col)
```

```python
def feature_engineering(X, y):
    print("=" * 50)
    print("FEATURE ENGINEERING")
    print("=" * 50)

    # Feature importance using Random Forest
    rf_temp = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_temp.fit(X, y)

    # Get feature importance
    feature_importance = pd.DataFrame({
        'feature': X.columns,
        'importance': rf_temp.feature_importances_
    }).sort_values('importance', ascending=False)

    print("Top 10 most important features:")
    print(feature_importance.head(10))

    # Visualize feature importance
    plt.figure(figsize=(10, 6))
    sns.barplot(data=feature_importance.head(15), x='importance', y='feature')
    plt.title('Top 15 Feature Importance')
    plt.tight_layout()
    plt.show()

    # Select top features (you can adjust this threshold)
    top_features = feature_importance.head(15)['feature'].tolist()
    X_selected = X[top_features]

    print(f"Selected {len(top_features)} features out of {len(X.columns)}")

    return X_selected, feature_importance

X_selected, feat_importance = feature_engineering(X, y)
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

```

```python
# ========================================
# COMPREHENSIVE SMOTE VARIANTS COMPARISON
# ========================================

print(" TESTING DIFFERENT SMOTE VARIANTS FOR THYROID CANCER DATASET")
print("=" * 60)

# Define SMOTE variants to test
smote_variants = {
    'Original SMOTE': SMOTE(random_state=42),
    'Borderline SMOTE-1': BorderlineSMOTE(random_state=42, kind='borderline-1'),
    'Borderline SMOTE-2': BorderlineSMOTE(random_state=42, kind='borderline-2'),
    'SVM SMOTE': SVMSMOTE(random_state=42),
    'ADASYN': ADASYN(random_state=42, n_neighbors=5)
}

# Store results for comparison
smote_results = {}
smote_datasets = {}

print(f"Original class distribution: {Counter(y_train)}")
print()

# Test each SMOTE variant
for smote_name, smote_method in smote_variants.items():
    print(f" Testing {smote_name}...")
    
    try:
        # Apply SMOTE
        X_train_smote, y_train_smote = smote_method.fit_resample(X_train, y_train)
        
        # Store the resampled data
        smote_datasets[smote_name] = (X_train_smote, y_train_smote)
        
        # Print class distribution after SMOTE
        class_dist = Counter(y_train_smote)
        print(f"   Class distribution after {smote_name}: {class_dist}")
        print(f"   Total samples: {len(y_train_smote)}")
        
        # Quick evaluation with a simple Random Forest
        from sklearn.ensemble import RandomForestClassifier
        from sklearn.preprocessing import StandardScaler
        from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
        
        rf_temp = RandomForestClassifier(n_estimators=100, random_state=42)
        
        # Scale the data
        scaler_temp = StandardScaler()
        X_train_scaled_temp = scaler_temp.fit_transform(X_train_smote)
        X_test_scaled_temp = scaler_temp.transform(X_test)
        
        # Train and predict
        rf_temp.fit(X_train_scaled_temp, y_train_smote)
        y_pred_temp = rf_temp.predict(X_test_scaled_temp)
        
        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred_temp)
        f1 = f1_score(y_test, y_pred_temp, average='weighted')
        precision = precision_score(y_test, y_pred_temp, average='weighted')
        recall = recall_score(y_test, y_pred_temp, average='weighted')
        
        smote_results[smote_name] = {
            'Accuracy': accuracy,
            'F1-Score': f1,
            'Precision': precision,
            'Recall': recall,
            'Samples': len(y_train_smote)
        }
        
        print(f"   Quick RF Evaluation - Accuracy: {accuracy:.4f}, F1: {f1:.4f}")
        
    except Exception as e:
        print(f"   Error with {smote_name}: {e}")
        smote_results[smote_name] = {
            'Accuracy': 0,
            'F1-Score': 0,
            'Precision': 0,
            'Recall': 0,
            'Samples': 0
        }
    
    print()

print(" SMOTE VARIANT TESTING COMPLETED!")
print("=" * 60)
```

```python
# ========================================
# SMOTE RESULTS ANALYSIS & VISUALIZATION
# ========================================

# Convert results to DataFrame for easy analysis
smote_comparison_df = pd.DataFrame(smote_results).T
print(" COMPREHENSIVE SMOTE VARIANT COMPARISON")
print("=" * 60)
print(smote_comparison_df.round(4))

# Find the best SMOTE variant
best_smote_method = smote_comparison_df['Accuracy'].idxmax()
best_accuracy = smote_comparison_df.loc[best_smote_method, 'Accuracy']
print(f"\n BEST SMOTE VARIANT: {best_smote_method}")
print(f" Best Accuracy: {best_accuracy:.4f}")

# Visualize the comparison
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Accuracy Comparison
axes[0, 0].barh(smote_comparison_df.index, smote_comparison_df['Accuracy'], color='skyblue')
axes[0, 0].set_title('Accuracy Comparison')
axes[0, 0].set_xlabel('Accuracy')
axes[0, 0].grid(True, alpha=0.3)

# 2. F1-Score Comparison
axes[0, 1].barh(smote_comparison_df.index, smote_comparison_df['F1-Score'], color='lightgreen')
axes[0, 1].set_title('F1-Score Comparison')
axes[0, 1].set_xlabel('F1-Score')
axes[0, 1].grid(True, alpha=0.3)

# 3. Precision vs Recall
axes[1, 0].scatter(smote_comparison_df['Precision'], smote_comparison_df['Recall'], 
                   s=100, alpha=0.7, c=range(len(smote_comparison_df)), cmap='viridis')
for i, method in enumerate(smote_comparison_df.index):
    axes[1, 0].annotate(method, 
                        (smote_comparison_df.loc[method, 'Precision'], 
                         smote_comparison_df.loc[method, 'Recall']),
                        xytext=(5, 5), textcoords='offset points', fontsize=8)
axes[1, 0].set_title('Precision vs Recall')
axes[1, 0].set_xlabel('Precision')
axes[1, 0].set_ylabel('Recall')
axes[1, 0].grid(True, alpha=0.3)

# 4. Sample Count
axes[1, 1].barh(smote_comparison_df.index, smote_comparison_df['Samples'], color='coral')
axes[1, 1].set_title('Generated Sample Count')
axes[1, 1].set_xlabel('Number of Samples')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Show detailed metrics for top 3 performers
print("\n TOP 3 SMOTE VARIANTS BY ACCURACY:")
print("=" * 50)
top_3 = smote_comparison_df.sort_values('Accuracy', ascending=False).head(3)
for i, (method, metrics) in enumerate(top_3.iterrows(), 1):
    print(f"{i}. {method}:")
    print(f"   Accuracy: {metrics['Accuracy']:.4f}")
    print(f"   F1-Score: {metrics['F1-Score']:.4f}")
    print(f"   Precision: {metrics['Precision']:.4f}")
    print(f"   Recall: {metrics['Recall']:.4f}")
    print()

# Get the best SMOTE method and data for final training
best_smote_data = smote_datasets[best_smote_method]
X_train_resampled, y_train_resampled = best_smote_data

print(f" Using {best_smote_method} for final model training")
print(f" Final class distribution: {Counter(y_train_resampled)}")

# Plot final distribution
plt.figure(figsize=(8, 5))
sns.countplot(x=y_train_resampled, palette='Set2')
plt.title(f'Class Distribution After {best_smote_method}')
plt.xlabel('Class')
plt.ylabel('Count')
plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.show()
```

```python

def train_and_tune_models_with_knn(X_train, X_test, y_train, y_test):
    print("=" * 50)
    print("MODEL TRAINING AND TUNING (WITH KNN + BOOSTING ALGORITHMS)")
    print("=" * 50)

    models = {}

    # 1. Random Forest with GridSearch
    print("Training Random Forest...")
    rf_params = {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    rf = RandomForestClassifier(random_state=42)
    rf_grid = GridSearchCV(rf, rf_params, cv=5, scoring='accuracy', n_jobs=-1)
    rf_grid.fit(X_train, y_train)
    models['Random Forest'] = rf_grid.best_estimator_
    print(f"Best RF params: {rf_grid.best_params_}")

    # 2. K-Nearest Neighbors with GridSearch
    print("Training K-Nearest Neighbors...")
    knn_params = {
        'n_neighbors': [3, 5, 7, 9, 11, 15, 21],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan', 'minkowski'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
        'p': [1, 2]  # p=1 for manhattan, p=2 for euclidean
    }

    knn = KNeighborsClassifier()
    knn_grid = GridSearchCV(knn, knn_params, cv=5, scoring='accuracy', n_jobs=-1)
    knn_grid.fit(X_train, y_train)
    models['K-Nearest Neighbors'] = knn_grid.best_estimator_
    print(f"Best KNN params: {knn_grid.best_params_}")


    # 4. Logistic Regression
    print("Training Logistic Regression...")
    lr_params = {
        'C': [0.1, 1, 10, 100],
        'penalty': ['l1', 'l2'],
        'solver': ['liblinear']
    }

    lr = LogisticRegression(random_state=42, max_iter=1000)
    lr_grid = GridSearchCV(lr, lr_params, cv=5, scoring='accuracy', n_jobs=-1)
    lr_grid.fit(X_train, y_train)
    models['Logistic Regression'] = lr_grid.best_estimator_
    print(f"Best LR params: {lr_grid.best_params_}")

    # 5. SVM
    print("Training SVM...")
    svm_params = {
        'C': [0.1, 1, 10],
        'kernel': ['rbf', 'linear'],
        'gamma': ['scale', 'auto']
    }

    svm = SVC(random_state=42)
    svm_grid = GridSearchCV(svm, svm_params, cv=5, scoring='accuracy', n_jobs=-1)
    svm_grid.fit(X_train, y_train)
    models['SVM'] = svm_grid.best_estimator_
    print(f"Best SVM params: {svm_grid.best_params_}")

    # 6. Decision Tree with GridSearch
    print("Training Decision Tree...")
    dt_params = {
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 5, 10, 15, 20],
        'min_samples_split': [2, 5, 10, 15],
        'min_samples_leaf': [1, 2, 5, 10],
        'max_features': ['auto', 'sqrt', 'log2', None]
    }

    dt = DecisionTreeClassifier(random_state=42)
    dt_grid = GridSearchCV(dt, dt_params, cv=5, scoring='accuracy', n_jobs=-1)
    dt_grid.fit(X_train, y_train)
    models['Decision Tree'] = dt_grid.best_estimator_
    print(f"Best DT params: {dt_grid.best_params_}")

 

    # 8. LightGBM with GridSearch
    print("Training LightGBM...")
    try:
        from lightgbm import LGBMClassifier

        lgb_params = {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 5, 7, -1],
            'learning_rate': [0.05, 0.1, 0.2],
            'num_leaves': [31, 50, 100],
            'subsample': [0.8, 0.9, 1.0],
            'colsample_bytree': [0.8, 0.9, 1.0]
        }

        lgb = LGBMClassifier(random_state=42, verbosity=-1)
        lgb_grid = GridSearchCV(lgb, lgb_params, cv=5, scoring='accuracy', n_jobs=-1)
        lgb_grid.fit(X_train, y_train)
        models['LightGBM'] = lgb_grid.best_estimator_
        print(f"Best LGB params: {lgb_grid.best_params_}")

    except ImportError:
        print("LightGBM not installed. Installing...")
        import subprocess
        subprocess.check_call(['pip', 'install', 'lightgbm'])
        from lightgbm import LGBMClassifier

        lgb_params = {
            'n_estimators': [100, 200, 300],
            'max_depth': [3, 5, 7, -1],
            'learning_rate': [0.05, 0.1, 0.2],
            'num_leaves': [31, 50, 100],
            'subsample': [0.8, 0.9, 1.0],
            'colsample_bytree': [0.8, 0.9, 1.0]
        }

        lgb = LGBMClassifier(random_state=42, verbosity=-1)
        lgb_grid = GridSearchCV(lgb, lgb_params, cv=5, scoring='accuracy', n_jobs=-1)
        lgb_grid.fit(X_train, y_train)
        models['LightGBM'] = lgb_grid.best_estimator_
        print(f"Best LGB params: {lgb_grid.best_params_}")

    
    print(f"\n TRAINING COMPLETED!")
    print(f"Total models trained: {len(models)}")
    print(f"Models: {list(models.keys())}")

    return models

# Train models with all algorithms including the new boosting methods
trained_models_with_knn = train_and_tune_models_with_knn(
    X_train_scaled, X_test_scaled, y_train_resampled, y_test
)
```

```python
from sklearn.metrics import log_loss

def evaluate_models_with_knn(models, X_train, X_test, y_train, y_test):
    print("=" * 50)
    print("MODEL EVALUATION (INCLUDING KNN)")
    print("=" * 50)

    results = {}

    for name, model in models.items():
        # Cross-validation score
        cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')

        # Train and test predictions
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

        # Probabilities for log loss
        try:
            y_test_proba = model.predict_proba(X_test)
            loss = log_loss(y_test, y_test_proba)
        except:
            loss = float('nan')  # if not available

        # Calculate accuracies
        train_accuracy = accuracy_score(y_train, y_train_pred)
        test_accuracy = accuracy_score(y_test, y_test_pred)

        results[name] = {
            'CV Mean': cv_scores.mean(),
            'CV Std': cv_scores.std(),
            'Train Accuracy': train_accuracy,
            'Test Accuracy': test_accuracy,
            'Log Loss': loss
        }

        print(f"\n{name}:")
        print(f"  Cross-validation: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
        print(f"  Train Accuracy: {train_accuracy:.4f}")
        print(f"  Test Accuracy: {test_accuracy:.4f}")
        print(f"  Log Loss: {loss:.4f}")

        if name in ['Random Forest', 'K-Nearest Neighbors']:
            print(f"\nDetailed Report for {name}:")
            print(classification_report(y_test, y_test_pred))

            plt.figure(figsize=(8, 6))
            cm = confusion_matrix(y_test, y_test_pred)
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
            plt.title(f'Confusion Matrix - {name}')
            plt.ylabel('True Label')
            plt.xlabel('Predicted Label')
            plt.show()

    # Create results DataFrame
    results_df = pd.DataFrame(results).T
    print("\n" + "=" * 50)
    print("SUMMARY OF ALL MODELS")
    print("=" * 50)
    print(results_df.round(4))

    # Find best model
    best_model_name = results_df['Test Accuracy'].idxmax()
    best_accuracy = results_df.loc[best_model_name, 'Test Accuracy']

    print(f"\nBest Model: {best_model_name}")
    print(f"Best Test Accuracy: {best_accuracy:.4f}")

    return results_df, models[best_model_name]
```
